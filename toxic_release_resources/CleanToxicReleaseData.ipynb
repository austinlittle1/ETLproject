## Cleaning EPA's Toxic Release Data

This code loads EPA Toxic Release Data between 2010-2017, extracts specific rows, and then combines all the data into one Pandas DataFrame

```python
import pandas as pd

# Load first data set 

ToxRelData = pd.read_csv('toxic-release-inventory/TRI_2010_US.csv')
ToxRelData = ToxRelData.iloc[:,[0,3,5,6,7,8,29,36,38,103,107]]
ToxRelData.head()
```

```python
# List of remaining toxic release data files
toxicList = ['TRI_2011_US.csv','TRI_2012_US.csv','TRI_2013_US.csv','TRI_2014_US.csv','TRI_2015_US.csv','TRI_2016_US.csv','TRI_2017_US.csv']

#Extract and Transform remaining files, concatenate them to original file
for toxicYear in toxicList:
    temp = pd.read_csv('toxic-release-inventory/' + toxicYear)
    temp = temp.iloc[:,[0,3,5,6,7,8,29,36,38,103,107]]
    ToxRelData = pd.concat([ToxRelData,temp])

ToxRelData
```
