```python
import pandas as pd
import numpy as np
```

```python
csv_file = "death_causes.csv"
cause_data = pd.read_csv(csv_file)
cause_data.head()
```

```python
cause_data.shape
```

```python
cause_data['Cause Name'].unique()
```

```python
cause_data['Year'].unique()
```

```python
cancer_data = cause_data.loc[(cause_data['Cause Name'] == 'Cancer')]
cancer_data = cancer_data.loc[cause_data['Year'] == 2011]
cancer_data.head()
```

```python
cancer_data.shape
```

```python
heart_data = cause_data.loc[(cause_data['Cause Name'] == 'Heart disease')]
heart_data = heart_data.loc[cause_data['Year'] == 2011]
heart_data.head()
```

```python
heart_data.shape
```

```python
influenza_data = cause_data.loc[(cause_data['Cause Name'] == 'Influenza and pneumonia')]
inf_data = influenza_data.loc[cause_data['Year'] == 2011]
inf_data.head()
```

```python
inf_data.shape
```

```python
air_file = "annual_aqi_by_county_2011.csv"
air_data = pd.read_csv(air_file)
air_data.head()
```

```python
air_data.columns
```

```python
air_data = air_data.loc[:, ['State', 'County', 'Year', '90th Percentile AQI']]
air_data.head()
```

```python
air_state = air_data.groupby('State')['90th Percentile AQI'].mean()
air_state = pd.DataFrame(air_state)
air_state.head()
```

```python
air_cancer = pd.merge(air_data, cancer_data, on=['State', 'Year'])
air_cancer.head()                    
```

```python
air_cancer_state = pd.merge(air_state, cancer_data, on=['State'])
air_cancer_state.head()
```
